### Import Libraries

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

### Getting the Data

```python
indofarma = pd.read_csv('INAF.csv')
kimiafarma = pd.read_csv('KAEF.csv')
phapros = pd.read_csv('PEHA.csv')
kalbefarma = pd.read_csv('KLBF.csv')
sidomuncul = pd.read_csv('SIDO.csv')
idx = pd.read_csv('JKSE.csv')
```

### Data Exploration

```python
indofarma['Open'].plot(label='Indofarma',figsize=(16,6),title='Open Price')
kalbefarma['Open'].plot(label='Kalbe Farma')
phapros['Open'].plot(label='Phapros')
kimiafarma['Open'].plot(label='Kimia Farma')
sidomuncul['Open'].plot(label='Sido Muncul')
plt.legend()
plt.ylabel('Rupiah')
```

```python
indofarma['Volume'].plot(label='Indofarma',figsize=(16,6),title='Volume Traded')
kalbefarma['Volume'].plot(label='Kalbe Farma')
phapros['Volume'].plot(label='Phapros')
kimiafarma['Volume'].plot(label='Kimia Farma')
sidomuncul['Volume'].plot(label='Sido Muncul')
plt.legend()
```

```python
indofarma['Total Traded'] = indofarma['Open']*indofarma['Volume']
kalbefarma['Total Traded'] = kalbefarma['Open']*kalbefarma['Volume']
phapros['Total Traded'] = phapros['Open']*phapros['Volume']
kimiafarma['Total Traded'] = kimiafarma['Open']*kimiafarma['Volume']
sidomuncul['Total Traded'] = sidomuncul['Open']*sidomuncul['Volume']
```

```python
indofarma['Total Traded'].plot(label='Indofarma',figsize=(16,6),title='Total Traded')
kalbefarma['Total Traded'].plot(label='Kalbe Farma')
phapros['Total Traded'].plot(label='Phapros')
kimiafarma['Total Traded'].plot(label='Kimia Farma')
sidomuncul['Total Traded'].plot(label='Sido Muncul')
plt.legend()
plt.ylabel('Rupiah')
```

```python
indofarma['MA20'] = indofarma['Open'].rolling(20).mean()
indofarma['MA50'] = indofarma['Open'].rolling(50).mean()
indofarma[['Open','MA20','MA50']].plot(label='Indofarma',figsize=(16,8),title='Indofarma')
plt.ylabel('Rupiah')
```

```python
kalbefarma['MA20'] = kalbefarma['Open'].rolling(20).mean()
kalbefarma['MA50'] = kalbefarma['Open'].rolling(50).mean()
kalbefarma[['Open','MA20','MA50']].plot(label='Kalbe Farma',figsize=(16,8),title='Kalbe Farma')
plt.ylabel('Rupiah')
```

```python
phapros['MA20'] = phapros['Open'].rolling(20).mean()
phapros['MA50'] = phapros['Open'].rolling(50).mean()
phapros[['Open','MA20','MA50']].plot(label='Phapros',figsize=(16,8),title='Phapros')
plt.ylabel('Rupiah')
```

```python
kimiafarma['MA20'] = kimiafarma['Open'].rolling(20).mean()
kimiafarma['MA50'] = kimiafarma['Open'].rolling(50).mean()
kimiafarma[['Open','MA20','MA50']].plot(label='Kimia Farma',figsize=(16,8),title='Kimia Farma')
plt.ylabel('Rupiah')
```

```python
sidomuncul['MA20'] = sidomuncul['Open'].rolling(20).mean()
sidomuncul['MA50'] = sidomuncul['Open'].rolling(50).mean()
sidomuncul[['Open','MA20','MA50']].plot(label='Sido Muncul',figsize=(16,8),title='Sido Muncul')
plt.ylabel('Rupiah')
```

```python
from pandas.plotting import scatter_matrix
pharma_comp = pd.concat([indofarma['Open'],kalbefarma['Open'],phapros['Open'],kimiafarma['Open'],sidomuncul['Open']],axis=1)
pharma_comp.columns = ['Indofarma Open','Kalbe Farma Open','Phapros Open','Kimia Farma Open','Sido Muncul Open']
scatter_matrix(pharma_comp,figsize=(12,12),alpha=0.2,hist_kwds={'bins':50})
```

### Daily Returns

```python
indofarma['returns'] = indofarma['Close'].pct_change(1)
kalbefarma['returns'] = kalbefarma['Close'].pct_change(1)
phapros['returns'] = phapros['Close'].pct_change(1)
kimiafarma['returns'] = kimiafarma['Close'].pct_change(1)
sidomuncul['returns'] = sidomuncul['Close'].pct_change(1)

idx['returns'] = idx['Close'].pct_change(1)
```

```python
indofarma['returns'].plot(kind='kde',label='Indofarma',figsize=(16,4))
kalbefarma['returns'].plot(kind='kde',label='Kalbe Farma')
phapros['returns'].plot(kind='kde',label='Phapros')
kimiafarma['returns'].plot(kind='kde',label='Kimia Farma')
sidomuncul['returns'].plot(kind='kde',label='Sido Muncul')
idx['returns'].plot(kind='kde',label='IDX Average')
plt.legend()
```

```python
box_df = pd.concat([indofarma['returns'],kalbefarma['returns'],phapros['returns'],kimiafarma['returns'],sidomuncul['returns'],idx['returns']],axis=1)
box_df.columns = ['Indofarma Returns',' Kalbe Farma Returns','Phapros Returns','Kimia Farma Returns','Sido Muncul Returns','IDX Average']
box_df.plot(kind='box',figsize=(16,10),colormap='jet')
```

### Cumulative Daily Returns

```python
indofarma['Cumulative Return'] = (1 + indofarma['returns']).cumprod()
kalbefarma['Cumulative Return'] = (1 + kalbefarma['returns']).cumprod()
phapros['Cumulative Return'] = (1 + phapros['returns']).cumprod()
kimiafarma['Cumulative Return'] = (1 + kimiafarma['returns']).cumprod()
sidomuncul['Cumulative Return'] = (1 + sidomuncul['returns']).cumprod()
idx['Cumulative Return'] = (1 + idx['returns']).cumprod()
```

```python
indofarma['Cumulative Return'].plot(label='Indofarma',figsize=(16,6),title='Cumulative Return')
kalbefarma['Cumulative Return'].plot(label='Kalbe Farma')
phapros['Cumulative Return'].plot(label='Phapros')
kimiafarma['Cumulative Return'].plot(label='Kimia Farma')
sidomuncul['Cumulative Return'].plot(label='Sido Muncul')
idx['Cumulative Return'].plot(label='IDX Average')
plt.legend()
```
