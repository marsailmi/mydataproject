### Import Libraries

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

### Getting the Data

```python
indofarma = pd.read_csv('INAF.csv')
kimiafarma = pd.read_csv('KAEF.csv')
phapros = pd.read_csv('PEHA.csv')
kalbefarma = pd.read_csv('KLBF.csv')
consumer = pd.read_csv('JKCONS.csv')
mining = pd.read_csv('JKMING.csv')
manufacturing = pd.read_csv('JKMNFG.csv')
finance = pd.read_csv('JKFINA.csv')
agriculture = pd.read_csv('JKAGRI.csv')
miscellaneous = pd.read_csv('JKMISC.csv')
infrastructure = pd.read_csv('JKINFA.csv')
basic = pd.read_csv('JKBIND.csv')
trade = pd.read_csv('JKTRAD.csv')
property = pd.read_csv('JKPROP.csv')
```

### Data Exploration

```python
from pandas.plotting import scatter_matrix
sectoral_comp = pd.concat([consumer['Open'],mining['Open'],manufacturing['Open'],finance['Open'],agriculture['Open'],miscellaneous['Open'],infrastructure['Open'],basic['Open'],trade['Open'],property['Open']],axis=1)
sectoral_comp.columns = ['Consumer','Mining','Manuf','Finance','Agri','Miscellan','Infra','Basic','Trade','Prop']
scatter_matrix(sectoral_comp,figsize=(16,16),alpha=0.2,hist_kwds={'bins':50})
```

### Daily Returns

```python
indofarma['returns'] = indofarma['Close'].pct_change(1)
kalbefarma['returns'] = kalbefarma['Close'].pct_change(1)
phapros['returns'] = phapros['Close'].pct_change(1)
kimiafarma['returns'] = kimiafarma['Close'].pct_change(1)
consumer['returns'] = consumer['Close'].pct_change(1)
mining['returns'] = mining['Close'].pct_change(1)
manufacturing['returns'] = manufacturing['Close'].pct_change(1)
finance['returns'] = finance['Close'].pct_change(1)
agriculture['returns'] = agriculture['Close'].pct_change(1)
miscellaneous['returns'] = miscellaneous['Close'].pct_change(1)
infrastructure['returns'] = infrastructure['Close'].pct_change(1)
basic['returns'] = basic['Close'].pct_change(1)
trade['returns'] = trade['Close'].pct_change(1)
property['returns'] = property['Close'].pct_change(1)
```

```python
indofarma['returns'].hist(bins=100,label='Indofarma',figsize=(16,8),alpha=0.5)
kalbefarma['returns'].hist(bins=100,label='Kalbe Farma',alpha=0.5)
phapros['returns'].hist(bins=100,label='Phapros',alpha=0.5)
kimiafarma['returns'].hist(bins=100,label='Kimia Farma',alpha=0.5)
consumer['returns'].hist(bins=100,label='Consumer Avg',alpha=0.5)
mining['returns'].hist(bins=100,label='Mining Avg',alpha=0.5)
manufacturing['returns'].hist(bins=100,label='Manuf Avg',alpha=0.5)
finance['returns'].hist(bins=100,label='Finance Avg',alpha=0.5)
agriculture['returns'].hist(bins=100,label='Agri Avg',alpha=0.5)
miscellaneous['returns'].hist(bins=100,label='Miscellan Avg',alpha=0.5)
infrastructure['returns'].hist(bins=100,label='Infra Avg',alpha=0.5)
basic['returns'].hist(bins=100,label='Basic Avg',alpha=0.5)
trade['returns'].hist(bins=100,label='Trade Avg',alpha=0.5)
property['returns'].hist(bins=100,label='Prop Avg',alpha=0.5)
plt.legend()
```

```python
indofarma['returns'].plot(kind='kde',label='Indofarma',figsize=(16,6))
kalbefarma['returns'].plot(kind='kde',label='Kalbe Farma')
phapros['returns'].plot(kind='kde',label='Phapros')
kimiafarma['returns'].plot(kind='kde',label='Kimia Farma')
consumer['returns'].plot(kind='kde',label='Consumer Avg')
mining['returns'].plot(kind='kde',label='Mining Avg')
manufacturing['returns'].plot(kind='kde',label='Manuf Avg')
finance['returns'].plot(kind='kde',label='Finance Avg')
agriculture['returns'].plot(kind='kde',label='Agri Avg')
miscellaneous['returns'].plot(kind='kde',label='Miscellan Avg')
infrastructure['returns'].plot(kind='kde',label='Infra Avg')
basic['returns'].plot(kind='kde',label='Basic Avg')
trade['returns'].plot(kind='kde',label='Trade Avg')
property['returns'].plot(kind='kde',label='Prop Avg')
plt.legend()
```

### Cumulative Daily Returns

```python
indofarma['Cumulative Return'] = (1 + indofarma['returns']).cumprod()
kalbefarma['Cumulative Return'] = (1 + kalbefarma['returns']).cumprod()
phapros['Cumulative Return'] = (1 + phapros['returns']).cumprod()
kimiafarma['Cumulative Return'] = (1 + kimiafarma['returns']).cumprod()
consumer['Cumulative Return'] = (1 + consumer['returns']).cumprod()
mining['Cumulative Return'] = (1 + mining['returns']).cumprod()
manufacturing['Cumulative Return'] = (1 + manufacturing['returns']).cumprod()
finance['Cumulative Return'] = (1 + finance['returns']).cumprod()
agriculture['Cumulative Return'] = (1 + agriculture['returns']).cumprod()
miscellaneous['Cumulative Return'] = (1 + miscellaneous['returns']).cumprod()
infrastructure['Cumulative Return'] = (1 + infrastructure['returns']).cumprod()
basic['Cumulative Return'] = (1 + basic['returns']).cumprod()
trade['Cumulative Return'] = (1 + trade['returns']).cumprod()
property['Cumulative Return'] = (1 + property['returns']).cumprod()
```

```python
indofarma['Cumulative Return'].plot(label='Indofarma',figsize=(16,8),title='Cumulative Return')
kalbefarma['Cumulative Return'].plot(label='Kalbe Farma')
phapros['Cumulative Return'].plot(label='Phapros')
kimiafarma['Cumulative Return'].plot(label='Kimia Farma')
consumer['Cumulative Return'].plot(label='Consumer Avg')
mining['Cumulative Return'].plot(label='Mining Avg')
manufacturing['Cumulative Return'].plot(label='Manuf Avg')
finance['Cumulative Return'].plot(label='Finance Avg')
agriculture['Cumulative Return'].plot(label='Agri Avg')
miscellaneous['Cumulative Return'].plot(label='Miscellan Avg')
infrastructure['Cumulative Return'].plot(label='Infra Avg')
basic['Cumulative Return'].plot(label='Basic Avg')
trade['Cumulative Return'].plot(label='Trade Avg')
property['Cumulative Return'].plot(label='Prop Avg')
plt.legend()
```
